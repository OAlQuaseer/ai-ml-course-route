1. Write a function that takes a list of strings and a number K. The function should
return a new list containing only the strings that have a length greater than K.
 Input: list=["apple", "pie", "banana"], K=3
 Output: ['apple', 'banana']

```python
def filter_strings_by_length(strings, k):
    return [s for s in strings if len(s) > k]

input_list = ["apple", "pie", "banana"]
k = 3
print(filter_strings_by_k(input_list, k))
```

2. Write a function named area() that calculates the area of either a "circle" or a
"rectangle". The function should take the shape name as the first argument,
followed by the necessary dimensions (radius for circle; length and width for
rectangle).
 Example Input: area("circle", 5) OR area("rectangle", 4, 6)
 Example Output: Circle Area: 78.5, Rectangle Area: 24




```python
def area(shape, *dimensions):
    if shape.lower() == "circle":
        radius = dimensions[0]
        result = 3.14 * (radius ** 2)
        print(f"Circle Area: {result}")
    elif shape.lower() == "rectangle":
        length, width = dimensions
        result = length * width
        print(f"Rectangle Area: {result}")

area("circle", 5)
area("rectangle", 4, 6)
```

3. Write a program that takes an email address as input and splits it into two parts: the
"username" and the "domain".
 Example Input: support@google.com
 Example Output: Username: support, Domain: google.com




```python
email_address = input("Enter your email address: ")
def email_address_parser(email_address):
    username, domain = email_address.split('@')
    return username, domain

username, domain = email_address_parser(email_address)
print(f"Username: {username}, Domain: {domain}")
```

4. Write a program to find the sum of all numbers in a list. ensure
the program ignores any elements in the list that are not numbers (like strings or
booleans).
 Example Input: [10,"apple", 20, True, 5]
 Example Output: Total Sum: 35

```python
input = [10, "apple", 20, True, 5]
def sum_only_numbers(data_list):
    total = 0
    for item in data_list:
        if type(item) in (int, float):
            total += item
    return total

print(f"Total Sum: {sum_only_numbers(input)}")
```

5. Write a Python class Restaurant with attributes like menu_items, book_table, and
customer_orders, and methods like add_item_to_menu, book_tables, and
customer_order. (Note: Use dictionaries and lists to store the data.)
Perform the following tasks now:
 Now add items to the menu.
 Make table reservations.
 Take customer orders.
 Print the menu.

 Print table reservations.
 Print customer orders.

```python
class Restaurant:
    def __init__(self):
        # Initialize menu_items as an empty dictionary
        self.menu_items = {}
        
        # Initialize book_table with fixed 12 tables, all empty (None)
        # Structure: {1: None, 2: None, ..., 12: None}
        self.book_table = {i: None for i in range(1, 13)}
        
        # Initialize customer_orders as an empty list
        self.customer_orders = []

    def add_item_to_menu(self, item, price):
        """Add an item and its price to the menu."""
        self.menu_items[item] = price
        print(f"Added {item} to menu at ${price}")

    def add_reservation(self, table_number, customer_name, time):
        """Book a specific table if it is available."""
        if table_number < 1 or table_number > 12:
            print(f"Error: Table {table_number} does not exist (1-12 only).")
            return

        if self.book_table[table_number] is not None:
            print(f"Error: Table {table_number} is already occupied.")
        else:
            self.book_table[table_number] = {'name': customer_name, 'time': time}
            print(f"Table {table_number} successfully booked for {customer_name} at {time}.")

    def add_customer_order(self, table_number, order_list):
        """Take an order for a specific table."""
        # Check if table has a reservation
        if self.book_table[table_number] is None:
             print(f"Warning: Taking order for Table {table_number} but it has no reservation. cancelling the order....")

        # Validate items exist in menu
        valid_items = []
        for item in order_list:
            if item in self.menu_items:
                valid_items.append(item)
            else:
                print(f"Item '{item}' not found in menu. Skipping.")
        
        if valid_items:
            order_entry = {
                'table_number': table_number,
                'items': valid_items,
                'status': 'Pending'
            }
            self.customer_orders.append(order_entry)
            print(f"Order placed for Table {table_number}: {valid_items}")

    def print_menu(self):
        print("\n--- MENU ---")
        for item, price in self.menu_items.items():
            print(f"{item}: ${price}")

    def print_reservations(self):
        print("\n--- RESERVATIONS ---")
        for table, valid in self.book_table.items():
            if valid:
                print(f"Table {table}: {valid['name']} at {valid['time']}")

    def print_orders(self):
        print("\n--- ACTIVE ORDERS ---")
        for order in self.customer_orders:
            print(f"Table {order['table_number']}: {', '.join(order['items'])} ({order['status']})")



r = Restaurant()

# 1. Add items
r.add_item_to_menu("Burger", 12.99)
r.add_item_to_menu("Pizza", 15.50)
r.add_item_to_menu("Salad", 8.00)

# 2. Book tables
r.add_reservation(5, "Omar", "19:00")
r.add_reservation(2, "Ali", "20:00")
r.add_reservation(5, "Ahmed", "21:00")

# 3. Take orders
r.add_customer_order(5, ["Burger", "Salad", "Lobster"])

# 4. Print everything
r.print_menu()
r.print_reservations()
r.print_orders()
```
