1. Write a function that takes a list of strings and a number K. The function should
return a new list containing only the strings that have a length greater than K.
 Input: list=["apple", "pie", "banana"], K=3
 Output: ['apple', 'banana']

```python
def filter_strings_by_length(strings, k):
    return [s for s in strings if len(s) > k]

input_list = ["apple", "pie", "banana"]
k = 3
print(filter_strings_by_k(input_list, k))
```

2. Write a function named area() that calculates the area of either a "circle" or a
"rectangle". The function should take the shape name as the first argument,
followed by the necessary dimensions (radius for circle; length and width for
rectangle).
 Example Input: area("circle", 5) OR area("rectangle", 4, 6)
 Example Output: Circle Area: 78.5, Rectangle Area: 24




```python
def area(shape, *dimensions):
    if shape.lower() == "circle":
        radius = dimensions[0]
        result = 3.14 * (radius ** 2)
        print(f"Circle Area: {result}")
    elif shape.lower() == "rectangle":
        length, width = dimensions
        result = length * width
        print(f"Rectangle Area: {result}")

area("circle", 5)
area("rectangle", 4, 6)
```

3. Write a program that takes an email address as input and splits it into two parts: the
"username" and the "domain".
 Example Input: support@google.com
 Example Output: Username: support, Domain: google.com




```python
email_address = input("Enter your email address: ")
def email_address_parser(email_address):
    username, domain = email_address.split('@')
    return username, domain

username, domain = email_address_parser(email_address)
print(f"Username: {username}, Domain: {domain}")
```

4. Write a program to find the sum of all numbers in a list. ensure
the program ignores any elements in the list that are not numbers (like strings or
booleans).
 Example Input: [10,"apple", 20, True, 5]
 Example Output: Total Sum: 35

```python
input = [10, "apple", 20, True, 5]
def sum_only_numbers(data_list):
    total = 0
    for item in data_list:
        if type(item) in (int, float):
            total += item
    return total

print(f"Total Sum: {sum_only_numbers(input)}")
```

5. Write a Python class Restaurant with attributes like menu_items, book_table, and
customer_orders, and methods like add_item_to_menu, book_tables, and
customer_order. (Note: Use dictionaries and lists to store the data.)
Perform the following tasks now:
 Now add items to the menu.
 Make table reservations.
 Take customer orders.
 Print the menu.

 Print table reservations.
 Print customer orders.

```python
class Restaurant:
    def __init__(self):
        # Initialize menu_items as an empty dictionary
        self.menu_items = {}
        
        # Initialize book_table with fixed 12 tables, all empty (None)
        # Structure: {1: None, 2: None, ..., 12: None}
        self.book_table = {i: None for i in range(1, 13)}
        
        # Initialize customer_orders as an empty list
        self.customer_orders = []

    def add_item_to_menu(self, item, price):
        """Add an item and its price to the menu."""
        self.menu_items[item] = price
        print(f"Added {item} to menu at ${price}")

    def add_reservation(self, table_number, customer_name, time):
        """Book a specific table if it is available."""
        if table_number < 1 or table_number > 12:
            print(f"Error: Table {table_number} does not exist (1-12 only).")
            return

        if self.book_table[table_number] is not None:
            print(f"Error: Table {table_number} is already occupied.")
        else:
            self.book_table[table_number] = {'name': customer_name, 'time': time}
            print(f"Table {table_number} successfully booked for {customer_name} at {time}.")

    def add_customer_order(self, table_number, order_list):
        """Take an order for a specific table."""
        # Check if table has a reservation
        if self.book_table[table_number] is None:
             print(f"Warning: Taking order for Table {table_number} but it has no reservation. cancelling the order....")

        # Validate items exist in menu
        valid_items = []
        for item in order_list:
            if item in self.menu_items:
                valid_items.append(item)
            else:
                print(f"Item '{item}' not found in menu. Skipping.")
        
        if valid_items:
            order_entry = {
                'table_number': table_number,
                'items': valid_items
            }
            self.customer_orders.append(order_entry)
            print(f"Order placed for Table {table_number}: {valid_items}")

    def print_menu(self):
        print("\n--- MENU ---")
        for item, price in self.menu_items.items():
            print(f"{item}: ${price}")

    def print_reservations(self):
        print("\n--- RESERVATIONS ---")
        for table, valid in self.book_table.items():
            if valid:
                print(f"Table {table}: {valid['name']} at {valid['time']}")

    def print_orders(self):
        print("\n--- ACTIVE ORDERS ---")
        for order in self.customer_orders:
            print(f"Table {order['table_number']}: {', '.join(order['items'])}")



r = Restaurant()

# 1. Add items
r.add_item_to_menu("Burger", 12.99)
r.add_item_to_menu("Pizza", 15.50)
r.add_item_to_menu("Salad", 8.00)

# 2. Book tables
r.add_reservation(5, "Omar", "19:00")
r.add_reservation(2, "Ali", "20:00")
r.add_reservation(5, "Ahmed", "21:00")

# 3. Take orders
r.add_customer_order(5, ["Burger", "Salad", "Lobster"])

# 4. Print everything
r.print_menu()
r.print_reservations()
r.print_orders()
```

6. A Bike Rental System Let us start writing the class Customer that is composed of the following
attributes:
 name
 age
Here are some of the methods that you need to add to your class Customer;
 def __init__(self).
 Getter / setter for all attributes.
class MainBikeRental that is composed of the following attributes:
 List of bikes
 List of Rentals
Here are some of the methods that you need to add to your class MainBikeRental;
 def __init__(self).
 Getter / setter for all attributes.
 requestBike: Takes a request from the customer for the number of bikes and check if
the requested number is available or not if yes, function will deduct number_of_bikes
from shop stock else it will print wait 10 mins. And you have to make sure that the age
of the customer is over than 6 years.
 returnBike: Allows customers to return their bikes to the rental shop by taking
rentalTime, num_of_bikes as parameters. And calculate the total bill cost by multiplying
rentalTime * 40. Also, add the returned bikes to the stock store again
 addBike: adds bikes into the list
 totalCost: returns the total cost of bikes in the store
class Bike that is composed of the following attributes:
 name
 price
Here are some of the methods that you need to add to your class Bike;
 def __init__(self).
 Getter / setter for all attributes.

```python
class Customer:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def get_name(self):
        return self.name

    def set_name(self, name):
        self.name = name

    def get_age(self):
        return self.age

    def set_age(self, age):
        self.age = age

# I added here a unique id for each bike object to make the system more realistic 
class Bike:
    def __init__(self, bike_id, name, price):
        self.id = bike_id
        self.name = name
        self.price = price

    def get_id(self):
        return self.id

    def set_id(self, bike_id):
        self.id = bike_id

    def get_name(self):
        return self.name

    def set_name(self, name):
        self.name = name

    def get_price(self):
        return self.price

    def set_price(self, price):
        self.price = price


class MainBikeRental:
    def __init__(self):
        self.stock = []
        self.rentals = []

    
    def get_stock(self):
        return self.stock

    def set_stock(self, stock):
        self.stock = stock

    def get_rentals(self):
        return self.rentals

    def set_rentals(self, rentals):
        self.rentals = rentals

    def addBike(self, bike):
        """Adds a bike to the stock."""
        self.stock.append(bike)
        print(f"Bike added: {bike.get_name()} (ID: {bike.get_id()})")

    def totalCost(self):
        """Returns the total cost (value) of bikes in the store."""
        total = sum(bike.get_price() for bike in self.stock)
        return total

    def requestBike(self, customer, num_bikes):
        """Handles a bike rental request."""
        # 1. Check Customer Age
        if customer.get_age() <= 6:
            print(f"Rental denied. Customer {customer.get_name()} is too young (Age: {customer.get_age()}).")
            return

        # 2. Check Stock Availability
        if len(self.stock) < num_bikes:
            print(f"Rental denied. Only {len(self.stock)} bikes available, but {num_bikes} requested. Wait 10 mins.")
            return

        # 3. Process Rental
        rented_bikes = []
        rented_bike_ids = []
        
        # Pop specific bike objects from stock to ensure uniqueness logic
        for _ in range(num_bikes):
            bike = self.stock.pop(0) 
            rented_bikes.append(bike)
            rented_bike_ids.append(bike.get_id())
            

        # Record the rental
        rental_record = {
            'customer_name': customer,
            'bikes': rented_bikes # Storing the actual objects to restore them exactly later
        }
        self.rentals.append(rental_record)

        print(f"Rental successful for {customer.get_name()}: {num_bikes} bikes (IDs: {rented_bike_ids}) rented.")
        print(f"Bikes remaining in stock: {len(self.stock)}")

    def returnBike(self, customer_name, rentalTime):
        """Handles returning bikes and calculating the bill."""
        # 1. Find the rental record
        record_to_remove = None
        for record in self.rentals:
            if record['customer_name'].get_name() == customer_name:
                record_to_remove = record
                break
        
        if not record_to_remove:
            print(f"No active rental found for {customer_name}.")
            return

        # 2. Calculate Bill
        # Logic: rentalTime * 40 * number of bikes
        num_bikes = len(record_to_remove['bikes'])
        bill = rentalTime * 40 * num_bikes
        
        # 3. Restore Stock (Return the exact same bike objects)
        for bike in record_to_remove['bikes']:
            self.stock.append(bike)
            
        # 4. Remove from rentals
        self.rentals.remove(record_to_remove)

        print(f"Return successful for {customer_name}.")
        print(f"Returned {num_bikes} bikes. Total Bill: ${bill}")
        print(f"Bikes now in stock: {len(self.stock)}")



shop = MainBikeRental()

# 1. Add Bikes (Unique IDs)
b1 = Bike(101, "Mountain Bike", 500.0)
b2 = Bike(102, "Road Bike", 600.0)
b3 = Bike(103, "City Bike", 300.0)
b4 = Bike(104, "Kids Bike", 150.0)

shop.addBike(b1)
shop.addBike(b2)
shop.addBike(b3)
shop.addBike(b4)

print(f"Total Stock Value: ${shop.totalCost()}")

# 2. Customers
pro_rider = Customer("Alice", 25)
kid_rider = Customer("Timmy", 5)

# 3. Test Scenarios
print("\nTest 1: Age Check")
shop.requestBike(kid_rider, 1) # Should fail

print("\nTest 2: Successful Rental")
shop.requestBike(pro_rider, 2) # Should get 101, 102

print("\nTest 3: Stock Check")
shop.requestBike(Customer("LateUser", 30), 3) # Should fail (only 2 left)

print("\nTest 4: Return")
shop.returnBike("Alice", 3) # rentalTime = 3
# Bill should be 3 * 40 * 2 = 240
```
