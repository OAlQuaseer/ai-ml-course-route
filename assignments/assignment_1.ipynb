1. Ask the user for a string that contains both letters and numbers (e.g., "Py2th0on3"). Write a
program that separates them into two lists: one for the letters and one for the numbers. Calculate the
sum of the numbers found and join the letters to form a word.
• Example Input: "Py2th0on3"
• Example Output: Word: "Python", Sum: 5

```python
user_input = input("Enter any string: ")
numbers_list = [int(char) for char in user_input if char.isdigit()]
letters_list = [char for char in user_input if char.isalpha()]
print(f"Word: {''.join(letters_list)}, Sum: {sum(numbers_list)}")
```

2. Ask a user to enter a password. Check if it is valid based on these rules: Length must be between 8
and 16 characters, and it must contain at least one uppercase letter and one special character (e.g.,
@, #, $).
• Example Input: "Guest@123"
• Example Output: "Password is Secure."

```python


def is_valid_password(s):
    length_ok = len(s) >= 8 and len(s) <= 16
    has_upper = any(c.isupper() for c in s)
    has_special = any(not c.isalnum() for c in s)
    
    return length_ok and has_upper and has_special


user_input = input("Enter a password: ")

if is_valid_password(user_input):
    print("Password is Secure.")
else:
    print("Password is not Secure.")

```

3. A store gives discounts based on the total price: Under $50 (No discount), $50 to $100 (10%
discount), and Over $100 (20% discount). If the user is a "Member," they get an additional 5% off the
final calculated price.
• Example Input: Price: 100, Member: Yes
• Example Output: Final Total: $85.5 (Note: 100 - 10% = 90; then 90 - 5% = 85.5)

```python
def calculate_final_price(price, is_member):
    discount = 0
    
    if price >= 50 and price <= 100:
        discount = 0.1
    elif price > 100:
        discount = 0.2
    else:
        discount = 0
    
    price = price * (1 - discount)

    if is_member:
        return price * (1 - 0.05)
    else:
        return price

print(f"Final Total: ${calculate_final_price(100, True)}")


```

4. Ask the user to input two separate lists of names (comma-separated). Create a third list that
contains only the names that appear in both lists.
• Example Input: List 1: "Ali, Ben, Sara", List 2: "Ben, Joy, Ali"
• Example Output: Common names: ['Ali', 'Ben']

```python
def common_names(str1, str2):
    names_list1 = list(map(lambda x: x.strip(), str1.split(",")))
    names_list2 = list(map(lambda x: x.strip(), str2.split(",")))
    common_names_list = [name for name in names_list1 if name in names_list2]
    return common_names_list

print(f'Common names: {common_names("Ali, Ben, Sara", "Ben, Joy, Ali")}')
```

5. Create a dictionary where keys are student names and values are lists of 3 test scores. Write a
program that calculates the average for each student. If the average is > 90, print the student's name
followed by "Honor Roll."
• Example Input: {'Sam': [95, 90, 92]}
• Example Output: Sam: Average 92.3 - Honor Roll

```python
students = {
    'Sam': [95, 90, 92],
    'Eva': [85, 91, 89],
    'Liam': [78, 85, 88],
    'Olivia': [92, 88, 95],
    'Noah': [80, 82, 84]
}

def calculate_students_avgs(students_dict):
    students_averages = {}
    for name, scores in students_dict.items():
        students_averages[name] = sum(scores) / len(scores)
        if students_averages[name] > 90:
            print(f'{name}: Average {students_averages[name]:.2f} - Honor Roll')
    return students_averages

students_averages = calculate_students_avgs(students)
print(students_averages)
```

7. Ask for a multi-digit number. Sum its digits. If the result is more than one digit, keep summing the
digits until you get a single-digit result.
• Example Input: 98
• Example Output: Digital Root: 8 (9+8=17, 1+7=8)

```python
# Assuming here the user enters a numerical string always.
user_input = input("Enter a multi-digit number")

sum_of_digits = sum(map(int , user_input))
digital_root = ""
while True:
    if len(str(sum_of_digits)) == 1:
        digital_root = sum_of_digits
        break
    else:
        sum_of_digits = sum(map(int , str(sum_of_digits)))

print(f"Digital Root: {digital_root}")   

```

8. Given a list of numbers with many duplicates, return a list of tuples showing each unique element
and its frequency (how many times it appeared).
• Example Input: [1, 1, 2, 3, 3, 3, 4]
• Example Output: [(1, 2), (2, 1), (3, 3), (4, 1)]

```python
list_1 = [1, 1, 2, 3, 3, 3, 4]
elements_frequency = {}
for element in list_1:
    if element in elements_frequency:
        elements_frequency[element] += 1
    else:
        elements_frequency[element] = 1
output = list(elements_frequency.items()) 
print(output)
```

9. Ask the user for a number $n$. Calculate the factorial (n!) using a loop, and then check if the
resulting factorial is divisible by the sum of numbers from 1 to $n$.
• Example Input: 4
• Example Output: Factorial is 24, Sum is 10. 24 is not divisible by 10.

```python
num = int(input("Enter a number"))
sum = 0
factorial = 1
for i in range(1, num + 1):
    sum += i
    factorial *= i

is_factorial_divisible_by_sum = factorial % sum == 0
if is_factorial_divisible_by_sum:
    print(f"Factorial is {factorial}, Sum is {sum}. {factorial} is divisible by {sum}.")
else:
    print(f"Factorial is {factorial}, Sum is {sum}. {factorial} is not divisible by {sum}.")

```

10. Start with a dictionary of items and prices: {'apple': 0.5, 'bread': 2.0, 'milk': 1.5}. Allow the user to
"Add" or "Remove" items from a cart. If they add an item not in the store, print an error. Finally, print
the total bill.
• Example Input: Add 'apple', Add 'milk'
• Example Output: Total Bill: $2.0

```python
store_dict = {'apple': 0.5, 'bread': 2.0, 'milk': 1.5}

cart_dict = {}
total_bill = 0

while True:
    try:
        user_input = input("Enter the item name: ").strip().lower()
        if user_input == 'quit':
            break

        if user_input not in store_dict:
            raise ValueError("Item not found in the store.")
        
        quantity = int(input("Enter the quantity: "))
        
        if(cart_dict.get(user_input)):
            cart_dict[user_input] += quantity
        else:
            cart_dict[user_input] = quantity

        total_bill += store_dict[user_input] * quantity
        
    except ValueError as e:
        print(e)
    
    finally:
        print("Cart:")
        for item, quantity in cart_dict.items():
            print(f"{item}: {quantity}")
        print(f"Total bill: {total_bill}")

print("User finished shopping")
```

11. Check if a full sentence is a palindrome, ignoring spaces and capitalization.
• Example Input: "Race car"
• Example Output: True

```python

user_input = input('enter any sentence')

def is_palindrome(user_input):
    # we need first to clean and normalize the string 
    user_input = user_input.replace(" ", "").lower()

    is_palindrome = True

    # if the string empty then its palindrome
    if(len(user_input) ==0):
        return is_palindrome

    # I am using the two pointers approach here. 
    left_pointer = 0
    right_pointer = len(user_input) - 1

    while True:
        if user_input[left_pointer] != user_input[right_pointer]:
            is_palindrome = False
            break
        left_pointer +=1
        right_pointer -=1
        if(not(right_pointer > left_pointer)):
            break

    return is_palindrome

print(is_palindrome(user_input))


```

13. In a list of numbers, find all "peaks." A peak is a number that is greater than the number to its left
and the number to its right.
• Example Input: [1, 5, 2, 6, 3]
• Example Output: Peaks: [5, 6]

```python
list1 = [1, 5, 2, 6, 3]
peaks = []
for pointer in range(1, len(list1) - 1):
    if list1[pointer - 1] < list1[pointer] > list1[pointer + 1]:
        peaks.append(list1[pointer])
print(peaks)
```

14. Take a dictionary like {'A': 1, 'B': 2, 'C': 1} and flip it. If two keys have the same value, the new value
should be a list of those keys.
• Example Input: {'A': 1, 'B': 2, 'C': 1}
• Example Output: {1: ['A', 'C'], 2: ['B']}

```python
dict1 = {'A': 1, 'B': 2, 'C': 1}
output = {}
for key, value in dict1.items():
    if value not in output:
        output[value] = [key]
    else:
        output[value].append(key)
print(output)
```

15. Create a dictionary where the keys are the vowels found in a user's string and the values are lists
of the indices where those vowels appear.
• Example Input: "banana"
• Example Output: {'a': [1, 3, 5]}

```python
user_input = input("Enter a string: ").lower()
vowels = "aeiou"
dict1 = {}
for i,char in enumerate(user_input):
    if char in vowels:
        if char not in dict1:
            dict1[char] = [i]
        else:
            dict1[char].append(i)
print(dict1)
```
